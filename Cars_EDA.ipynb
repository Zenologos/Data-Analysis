```
 Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
 Car_Price
```

```
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
car_Horsepower
```

```
df1 = pd.DataFrame (JapaneseCars, columns = ['Company','Price'])
df2 = pd.DataFrame (GermanCars, columns = ['Company','Price'])
result = pd.concat([df1,df2])
```

```
result
```

```
df3 = pd.DataFrame (Car_Price, columns = ['Company','Price'])
df4 = pd.DataFrame (car_Horsepower, columns = ['Company','horsepower'])
MergedDF = pd.merge(df3, df4, on="Company")
```
